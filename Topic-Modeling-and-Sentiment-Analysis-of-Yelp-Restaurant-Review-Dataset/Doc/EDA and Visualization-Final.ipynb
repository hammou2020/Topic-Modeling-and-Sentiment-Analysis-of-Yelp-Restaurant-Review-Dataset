```python
documents = [(rid, t, star) for rid, t, star in zip(yelp_PN['review_id'], yelp_PN['text'], yelp_PN['stars'])]
```

```python
# Calculate polarity score and subjectivity score using TEXTBLOB
```

```python
from textblob import TextBlob

group = pd.DataFrame(documents)
group.rename(columns = {0: 'review_id', 1:'review_text', 2:'star'}, inplace = True)
group['polarity'] = [TextBlob(s).sentiment.polarity for s in group['review_text']]
group['subjectivity'] = [TextBlob(s).sentiment.subjectivity for s in group['review_text']]
group['word_list'] = [TextBlob(s).words.lemmatize() for s in group['review_text']]
print(list(group))
group.head()
```

```python
group['subjectivity'].describe()
```

```python
group.to_csv (r'rest_PN_sentiment.csv', index = None, header=True)
```

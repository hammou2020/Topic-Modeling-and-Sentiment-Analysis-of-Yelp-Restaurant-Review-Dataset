```python
import nltk

all_words = []

for (text, stars_x) in rev_star:
    for word in text.split():
        l = word.lower().replace('.', '').replace(',', '').replace('!', '')
        all_words.append(ps.stem(l))
        
all_words = nltk.FreqDist(all_words)
```

```python
# Get the most common 5000 words as features, if a word in a review is in this 5000 words,
# then return TRUE, otherwise return FALSE
words_lists = [item for (item,_) in list(all_words.most_common(5000))]

# Define a function to get lower case stemmed words, and if a word in a review is in this 5000 words, then return TRUE, otherwise return FALSE
stem = PorterStemmer()
def features(texts):
    words = set(texts.split())
    getf = {}
    for item in words:
        item = item.lower().replace('.', '').replace(',', '').replace('!', '')
        item = stem.stem(item)
        getf[item] = (item in words_lists)
    return getf

# Add star into the feature sets
featuresets = [(features(text), stars_x) for (text,stars_x) in rev_star]
print(featuresets[20])
```

```python
words_lists
```

```python
# For star >= 4, label as positive, others as negative
data = []
positive_number = 0
for item in range(len(featuresets)):
    if featuresets[item][1] >= 4:
        data.append((featuresets[item][0], 'positive'))
        positive_number = positive_number + 1
    else:
        data.append((featuresets[item][0], 'negative'))
print("Positive percent: ", positive_number / len(data)*100) # About 70% positive words
print("length of data: ", len(data)) # 71122
```

```python
# Training/ testing split, about 30% test data
training_set = data[:49000]
testing_set = data[49000:]
```

```python
# Try naive bayes, logistic regression and SVM models
from nltk.classify.scikitlearn import SklearnClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC

Naive_Bayes_clf = nltk.NaiveBayesClassifier.train(training_set)
print("Naive Bayes accuracy:", (nltk.classify.accuracy(Naive_Bayes_clf, testing_set))*100)

# Most imformative features: P(label) gives the probability that an input will receive each label, given no information about the input's features. 
# P(fname=fval|label) gives the probability that a given feature (fname) will receive a given value (fval), given that the label (label).
# Return a list of the 'most informative' features used by this classifier. For the purpose of this function, the informativeness of a 
# feature ``(fname,fval)`` is equal to the highest value of P(fname=fval|label), for any label, divided by the lowest value of P(fname=fval|label), for any label

Naive_Bayes_clf.show_most_informative_features(15)
# 51.38%

Logistic_Regression_classifier = SklearnClassifier(LogisticRegression())
Logistic_Regression_classifier.train(training_set)
print("Logistic_Regression_classifier accuracy:", (nltk.classify.accuracy(Logistic_Regression_classifier, testing_set))*100)
# 91.56%, dramatically outperform naive bayes

SVM_classifier = SklearnClassifier(LinearSVC())
SVM_classifier.train(training_set)
print("SVM_classifier accuracy:", (nltk.classify.accuracy(SVM_classifier, testing_set))*100)
# 90.29%, dramatically outperform naive bayes
```

```python
# Play with 5 kinds of models
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier

clf = LogisticRegression(random_state=0)
clf_A = GaussianNB()
clf_B = DecisionTreeClassifier(random_state=1)
clf_C = GradientBoostingClassifier(random_state=1)
clf_D = RandomForestClassifier(random_state=1)
```
